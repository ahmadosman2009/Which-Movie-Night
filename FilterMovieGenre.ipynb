```python
def filterGenre2(dataframe, filterByGenre):
    dataframe.reset_index(drop = True, inplace = True)
    row_list = []
    for i in dataframe.index:
        for genre in dataframe['genres'][i]:
#         for genre in dataframe['genres'][i].values.tolist():
            if genre == filterByGenre:
                row_list.append(i)
    df_new = dataframe.iloc[row_list]
    df_new = df_new.sort_values('wr', ascending=False)
    df_new = df_new[0:50].sample(n=20)
#     df_new.reset_index(drop = True, inplace = True)
    return df_new
```

```python
noNA = pd.read_csv('posters_18k_clean.csv')
filterGenre2(noNA, 'Comedy')
```
